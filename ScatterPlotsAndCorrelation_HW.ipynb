```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://www.andrew.cmu.edu/user/dvaroday/Math4AI/data/fertility-and-life-expectancy-2024.txt'
df = pd.read_table(url, sep='\t', engine='python')

display(df)
```

```python
## Q1.1

corr_coef_matrix = df.corr(numeric_only=True)
display(corr_coef_matrix)
```

```python
## Q1.2

plt.scatter(df.Fertility, df.Life_expectancy)
for i, txt in enumerate(df.Location):
    plt.annotate(txt, (df.Fertility[i],df.Life_expectancy[i]))
plt.subplots_adjust(bottom=0, top=1.5, left=0, right=1.5)
plt.xlabel('Fertility')
plt.ylabel('Life expectancy')
plt.title('Fertility vs. Life expect')
```

```python
## Q1.3

df_remove_one = df[df.Location!='Niger']
corr_coef_matrix = df_remove_one.corr(numeric_only=True)
display(corr_coef_matrix)
```
